```python
import sklearn
import pandas as pd
import numpy as np
import chardet
from csv import reader

from sklearn.model_selection import cross_val_score

from sklearn import tree
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
```

```python
#Read the data into a Pandas dataframe

attributes = []
for i in ["a","b","c","d","e","f","g"]:
    for j in ["1","2","3","4","5","6"]:
        attributes.append(i+j)
        
attributes.append("class")

data = pd.read_csv("../data/connect-4.data",names=attributes)
```

```python
data.tail()
```

```python
VALUE_MAP = {
    'x' :0,
    'o' :1,
    'b' :2,
    'win' : 0,
    'loss' : 1,
    'draw' : 2
}
```

```python
data = data.replace(VALUE_MAP)
```

```python
data.tail()
```

```python
x = data[data.columns[:-1]]
y = data['class']
```

#### Naive Bayes :

```python
naive_bayes_classifier = MultinomialNB()
```

```python
cv_score_naive_bayes = cross_val_score(naive_bayes_classifier, x, y, cv=10).mean()
print("Cross Validation score for Naive Bayes :",cv_score_naive_bayes)
```

#### Decision Tree :

```python
decision_tree_classifier = tree.DecisionTreeClassifier(min_samples_split=1000)
```

```python
cv_score_decision_tree = cross_val_score(decision_tree_classifier, x, y, cv=10).mean()
print("Cross Validation score for Decison Tree :",cv_score_decision_tree)
```
